**4. Identificacion de tipos de variables**
> Teniendo en cuenta las tablas anteriores es facil revisar que tipo de variables tenemos.


```python
variables_cualitativas_nominales = [
    "CodigoEstacion",
    "CodigoSensor",
    "NombreEstacion",
    "Departamento",
    "Municipio",
    "ZonaHidrografica",
    "DescripcionSensor",
    "UnidadMedida"
]

variables_cuantitativas_continuas = [
    "FechaObservacion",
    "ValorObservado",
    "Latitud",
    "Longitud"
]
```

**5. Identificacion de valores unicos para las variables nominales**
> Aqui podemos destacar que la variable "ZonaHidrografica" contiene valores atipicos como "Meta".

```python
for var in variables_cualitativas_nominales:
  print(f"\nValores únicos de {var}:\n")
  _ = [print(value) for value in df[var].unique()]
```

**6. Seleccion de variables a usar**

> Consideramos que las variables importantes son **FechaObservacion, ValorObservado, Latitud y Longitud**. Las demas no aportan informacion relevante para el analisis y/o son redundantes dada la posicion en coordenadas.




```python
# Eliminar las columnas
df_selected = df[["FechaObservacion", "ValorObservado", "Latitud", "Longitud"]]
# Verificar que las columnas han sido eliminadas
print(df_selected.head())
```

**7. Creacion de nuevas variables**
> Generamos la columna fecha, para ello lo calculamos a partir del string "FechaObservacion".


```python
# Aqui convertimos la fecha de observacion en string a formato unix time
df_selected["Fecha"] = pd.to_datetime(df_selected['FechaObservacion'], format="%m/%d/%Y %I:%M:%S %p").dt.strftime("%d/%m/%Y")
```

```python
# Eliminar las columnas
df_selected = df_selected[["Fecha", "ValorObservado", "Latitud", "Longitud"]]
# Verificar que las columnas han sido eliminadas
print(df_selected.head())
```

**8. Coeficientes de asimetria y curtosis**

```python
df_selected[["ValorObservado", "Latitud", "Longitud"]].skew()
```

> La asimetria de ValorObservado es positiva.

> La asimetria de Latitud y longitud es negativa.

```python
df_selected[["ValorObservado", "Latitud", "Longitud"]].kurt()
```

> La curtosis es leptocurtica en ValorObservado y Longitud

> La curtosis es mesocurtica en Latitud

**9. Graficos de cajas y bigotes**

```python
# gráficos de caja
df_selected.boxplot(column='ValorObservado')
plt.title("Gráficos de caja de la variable ValorObservado")
plt.show()

df_selected.boxplot(column='Latitud')
plt.title("Gráficos de caja de la variable Latitud")
plt.show()

df_selected.boxplot(column='Longitud')
plt.title("Gráficos de caja de la variable Longitud")
plt.show()
```

**10. Limpieza de datos**

> Aqui quitamos las observaciones que incluyen localizaciones atipicas

```python
df_filtered = df_selected[(df_selected["Longitud"] > -74.25) & (df_selected["Latitud"] > 4.3)]

print(len(df_filtered))
```

```python
# gráficos de caja
df_filtered.boxplot(column='Latitud')
plt.title("Gráficos de caja de la variable Latitud")
plt.show()

df_filtered.boxplot(column='Longitud')
plt.title("Gráficos de caja de la variable Longitud")
plt.show()
```

> Aqui eliminamos observaciones duplicadas y/o vacias

```python
df_filtered = df_filtered.dropna()
df_filtered = df_filtered.drop_duplicates()
```

**12. Creacion de variables: PrecipitacionDiaria**
> Primero vamos a agrupar por dias y luego calcularemos el promedio de las lluvias registradas por los sensores para ese dia.

```python
# Agrupar por fecha truncada al día y calcular el promedio de las medidas
df_agrupado = df_filtered.groupby(df_filtered['Fecha']).agg({'ValorObservado': 'mean'}).reset_index()
# Dimensiones del DataFrame (filas, columnas)
print(f"El dataframe tiene {df_agrupado.shape[0]} filas y {df_agrupado.shape[1]} columnas")
df_agrupado.head()
```

```python
df_agrupado.shape
```

```python
# Definir los umbrales
bins = [-0.1, 0, 0.1, 2.5, float('inf')]  # Umbrales: -0.1-0, 0-0.1, 0.1-2.5, >2.5
labels = ['No lluvia', 'Rastro lluvia', 'Lluvioso', 'Lluvia fuerte']  # Etiquetas correspondientes

# Crear una nueva columna con las etiquetas basadas en los umbrales
df_agrupado['PrecipitacionDiaria'] = pd.cut(df_agrupado['ValorObservado'], bins=bins, labels=labels, include_lowest=True)
df_agrupado.head()
```

```python
df_precipitacion = df_agrupado[["Fecha", "PrecipitacionDiaria"]]
df_precipitacion.head()
```
